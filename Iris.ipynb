```
#converting class into boolean as b=0 and g=1
for i in range(0,df.shape[0]):
  if df['Species'][i]=='Iris-setosa':
    df['Species'][i]=1;
  if df['Species'][i]=='Iris-versicolor':
    df['Species'][i]=2;
  if df['Species'][i]=='Iris-virginica':
    df['Species'][i]=3;
#typecasting as integer
#df['label']=df['label'].astype(int)d
```

```
matplotlib.style.use('ggplot')
fig = plt.figure(figsize=(15,8)) 
ax = fig.add_subplot(111, projection='3d',facecolor='#2f3e46')
ax.scatter(df['Species'], 
           df['SepalLengthCm'], 
           df['PetalLengthCm'], 
           s=20,
           c='#000000')
ax.view_init(30, 185)
ax.set_title('Scatter Plot')
ax.set_xlabel(df.columns[5],color='#ffffff')
ax.set_ylabel(df.columns[1],color='#ffffff')
ax.set_zlabel(df.columns[3],color='#ffffff')
plt.show()
```
